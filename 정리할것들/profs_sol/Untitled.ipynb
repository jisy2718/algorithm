```python
def find_min(*args):
    min_r = args[0]
    min_index = 0
    for idx, arg in enumerate(args):
        if arg < min_r:
            min_r = arg
            min_index = idx
            
    return min_index, min_r

    
def find_max(*args):
    max_r = args[0]
    max_index = 0
    for idx, arg in enumerate(args):
        if arg > max_r:
            max_r = arg
            max_index = idx
            
    return max_index, max_r

def sum_num(*args):
    result = 0
    for arg in args:
        result += arg
    return result

# 1. input 
dump = int(input())

arr = list(map(int,input().split()))

# 2. 평탄화가 완료될 경우의 높이
completed_height = sum_num(*arr)//100
#print(completed_height)
completed = False
completed_diff = None

# 2. min, max 찾아서 올기기
#i = 1
while dump > 0:
    #print(f"{i}-th iteration") ; i+=1
    max_index, max_value = find_max(*arr)
    min_index, min_value = find_min(*arr)
    #print(f"max_value :{max_value}, min_value :{min_value}")
    #print(f"arr : {arr}")
    # 끝나지 않은 경우
    if min_value < completed_height:
        arr[max_index] -= 1
        arr[min_index] += 1
        #print(f"1------------------")
        
    # 끝난 경우
    else:
        #print(f"2------------------")
        completed = True
        # 평탄화 결과가 평평한 경우
        if max_value == min_value:
            completed_diff = 0
            #print(f"2-1------------------")
            break
        
        # 평탄화 결과가 평평하지 않은 경우
        else:
            completed_diff = 1
            #print(f"2-2------------------")
            break
    dump -= 1
    
if completed == False:
    completed_diff = find_max(*arr)[1] - find_min(*arr)[1]

print(f"#{tc} {completed_diff}")
```

```python
def is_prime(n:int):
    if n == 1:
        return False
    elif n == 2:
        return True
    
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

T = int(input())
for tc in range(1,T+1):
    N = int(input())
    result = []
    length = N//2
    for num in range(2,length):
        
        # N 이 1이되면 인수분해 끝  
        if N == 1:
            break
        # 소수인 경우에만 인수분해 시도    
        if is_prime(num):
#             print(f"prime :{num}-------------------------")
            # 몇 번 나눌 수 있는지 count
            count = 0
            # 나머지가 있다는 것은 더 이상 해당 소수를 약수로 가지고 있지 않다는 것
            while (N % num == 0):
                count += 1
#                 print(f"count :{count}-------------------------")
                N //= num
            # 결과를 추가
            result.append(count)
#             print(f"result :{result}-------------------------")
            
            
    str_result = ' '.join(map(str,result))
#     print(f"str_result :{str_result}-------------------------")
    print(f"#{tc} "+ str_result)
```
