```python
T = 10
for tc in range(1,T+1):
    N = 100
    tc = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]
    
    max_sum = 0
    
    # 행의 합
    for i in range(N):
        cur_sum = 0
        for j in range(N):
            cur_sum += arr[i][j]
        if max_sum < cur_sum:
            max_sum = cur_sum
        
            
    # 열의 합
    for j in range(N):
        cur_sum = 0
        for i in range(N):
            cur_sum += arr[i][j]
        if max_sum < cur_sum:
            max_sum = cur_sum    
    
    # 대각선의 합
    cur_sum1 = 0
    cur_sum2 = 0
    for i in range(N):
        cur_sum1 += arr[i][i]
        cur_sum2 += arr[i][N-1-i]
    
    if max_sum < cur_sum1:
        max_sum = cur_sum1
    
    if max_sum < cur_sum2:
        max_sum = cur_sum2

    print(f"#{tc} {max_sum}")
```

# 교재 p17 연습문제 2
+ 10개의 정수 입력받아, 부분집합의 합이 0이 되는 것이 있는지 찾는 함수

```python
def partial_sum(arr:list):
    n = len(arr)
    # 부분집합 개수로 [0,1,0,0,1] 과 같이 부분집합을 표현
    # i의 비트모양이 부분집합의 모양 ( 부분집합에 포함되는 요소 )
    for i in range(1<<n):
        partial_sum = 0
        subset = []
        # arr의 각 자리가 위의 부분집합에서 1과 겹치는 경우들이 부분집합이다.
        for j in range(len(arr)):
            # i의 이진표현의 j번째 비트를 확인
            if i & (1<<j):
                partial_sum += arr[j]  # arr 의 j 번째 자리가 해당 i-th 부분집합에 들어가므로 더해준다.
                subset.append(arr[j])
        if partial_sum == 10:
            return (True, subset)
    return False
```

# 교재 p50 연습문제 3

+ ----------->
+  |-------> |
+  |         |
+  |         |
+  <---------           

```python
def snail_sort(arr:list):
    N = 5
    total = []
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            total.append(arr[i][j])
    total.sort()
    
    print(total)
    
 
    # 우하좌상
    dr = [0, 1, 0, -1]
    dc = [1, 0, -1, 0]
    direction = 0
    r = c = 0
    result = [[0]*N for _ in range(N)]
    
    num = 0
    while num < N*N:
        if 0 <= r < 5 and 0 <= c < 5 and not result[r][c] :
            result[r][c] = total[num]
            num += 1
   
        else:
            r = r - dr[direction]
            c = c - dc[direction]
            direction = ( direction + 1 ) % 4    
            
        
        r = r + dr[direction]
        c = c + dc[direction]
        
        for row in result:
            print(result)
        print("-"*10)
    return (result)
    
```
